#Waveforms. Amplitude vs Magnitude

```python
import librosa.display
plt.figure(figsize=(10,15))
librosa.display.waveplot(y2,sr=sr)
```

```python
librosa.display.waveplot(y2,sr=int(sr/2))
```

```python
librosa.display.waveplot(y2,sr=sr*2)
```

Amplitude of a wave is just the distance, positive or negative, from the equilibrium (zero in our case), and magnitude is the absolute value of the amplitude. In audio we sample the amplitude so our signal can have both positive and negative values.
