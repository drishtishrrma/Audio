```python
#import all the necessary libraries
from IPython.display import Audio
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Load Audio file
white_noise_file1 = "/content/audiocheck.net_whitenoise.wav"
white_noise_file2 = "/content/audiocheck.net_whitenoisegaussian.wav"
```

```python
white_noise1, _ = librosa.load(white_noise_file1)
white_noise2, _ = librosa.load(white_noise_file2)
```

```python
Audio(white_noise_file1)
```

```python
Audio(white_noise_file2)
```

```python
#visualizing the waveforms

plt.figure(figsize=(15, 17))

plt.subplot(5,1,1)
librosa.display.waveplot(white_noise1, alpha=0.5)
plt.title("Waveplot of White noise 1 Sample")

plt.subplot(5,1,2)
librosa.display.waveplot(white_noise2, alpha=0.5)
plt.title("Waveplot of White noise 2 Sample")
```

```python
FRAME_SIZE = 1024
HOP_LENGTH =128
#Calculating the amplitude envelope
def amplitude_envelope(signal, frame_size, hop_length):
  return np.array([max(signal[i:i+frame_size]) for i in range(0, signal.size, hop_length)])
```

#Amplitude Envelope

```python
#Amplitude Envelope for individual genre
ae_white_noise1 = amplitude_envelope(white_noise1, FRAME_SIZE, HOP_LENGTH)
ae_white_noise2 = amplitude_envelope(white_noise2, FRAME_SIZE, HOP_LENGTH)
```

```python
#visualizing Amplitude Envelope


frames = range(0, ae_white_noise1.size)
t = librosa.frames_to_time(frames, hop_length=HOP_LENGTH)

plt.figure(figsize=(15, 17))

plt.subplot(5,1,1)
librosa.display.waveplot(ae_white_noise1, alpha=0.5)
plt.plot(t, ae_white_noise1, color="r")
plt.title("Amplitude Envelope of White Noise 1 Sample")

plt.subplot(5,1,2)
librosa.display.waveplot(ae_white_noise2, alpha=0.5)
plt.plot(t, ae_white_noise2, color="r")
plt.title("Amplitude Envelope of White Noise 2 Sample")
```

#RMS Energy

```python
rms_white_noise1 = librosa.feature.rms(white_noise1, frame_length=FRAME_SIZE, hop_length=HOP_LENGTH)[0]
rms_white_noise2 = librosa.feature.rms(white_noise2, frame_length=FRAME_SIZE, hop_length=HOP_LENGTH)[0]
```

```python
#Visualizing RMS Energy

plt.figure(figsize=(15, 17))

plt.subplot(5,1,1)
librosa.display.waveplot(rms_white_noise1, alpha=0.5)
plt.plot(t, rms_white_noise1, color="r")
plt.title("RMS Energy of White Noise 1 Sample")

plt.subplot(5,1,2)
librosa.display.waveplot(rms_white_noise2, alpha=0.5)
plt.plot(t, rms_white_noise2, color="r")
plt.title("RMS Energy of White Noise 2 Sample")
```

#ZCR

```python
#Extracting zero crossing rate for each music genre song using Librosa 
zcr_white_noise1 = librosa.feature.zero_crossing_rate(white_noise1, frame_length=FRAME_SIZE, hop_length=HOP_LENGTH)[0]
zcr_white_noise2 = librosa.feature.zero_crossing_rate(white_noise2, frame_length=FRAME_SIZE, hop_length=HOP_LENGTH)[0]
```

```python
#Visualizing NORMALIZED Zero crossing rate (ZCR) of different music genres
plt.figure(figsize=(20, 20))

ax = plt.subplot(5, 1, 1)
librosa.display.waveplot(white_noise1, alpha=0.5)
plt.plot(t, zcr_white_noise1, color="r")
plt.ylim((-1, 1))
plt.title("White Noise 1")

plt.subplot(5, 1, 2)
librosa.display.waveplot(white_noise2, alpha=0.5)
plt.plot(t, zcr_white_noise2, color="g")
plt.ylim((-1, 1))
plt.title("White_noise2")
```
